```python
'''
I've made it this far without ever using matlab. I won't be stopped now.

% double click a function and press F1 in order to see the documentation
h = 0.2;    % step size
t = 0:h:2;  % time steps
w = zeros(length(t),1); 
w(1) = 0.5; % initial value
for i = 1:10
    w(i+1) = w(i)+h*( w(i)-t(i)^2+1 );
end
plot(t,w,'-o')
'''
import numpy as np
from matplotlib import pyplot
```

```python
N = 100
h = (4 * np.pi) / N
t = np.arange(start=0, stop=4*np.pi, step=h)
w = np.zeros(N)
x = np.zeros(N)
x[0] = -1
w[0] = 0
```

```python
for i in range(1, N-1):
    x[i+1] = x[i] + h * w[i-1]
    w[i+1] = w[i] + h * -np.sin(x[i-1])
```

```python
w_ = pyplot.plot(t, w, label="w")
x_ = pyplot.plot(t, x, label="x")
pyplot.xlabel('t')
pyplot.ylabel('w(blue), x(orange)')
pyplot.show()
```
